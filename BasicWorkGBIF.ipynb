```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('examples/GBIF_World.csv', sep = ",")
df.head()
```

```python
df. describe()
```

```python
x = df['year'].array
#Eliminación de NaN
x = x[np.logical_not(np.isnan(x))]
x = x.astype(int)
x
```

```python
intervalos = range(1860, max(x)+2) #calculamos los extremos de los intervalos
año = list(range(1860, 2021, 20))
plt.style.use('ggplot')
#plt.style.use('fivethirtyeight')
plt.hist(x=x, bins=intervalos, color='blue', rwidth=0.85)
plt.title('Entomophthorales reportados en GBIF')
plt.xlabel('Año')
plt.ylabel('Frecuencia')
plt.xticks(año,rotation=45)
plt.savefig("Histograma.png",dpi=300)
```

```python
# Para variables discretas
pd.value_counts(df['species'])
```

```python
100 * df['species'].value_counts() / len(df['species'])
```

```python
df_Mex = pd.read_csv('examples/GBIF_Mexico.csv', sep = ",")
```

```python
x_Mex = df_Mex['year'].array
#Eliminación de NaN
x_Mex = x_Mex[np.logical_not(np.isnan(x_Mex))]
x_Mex = x_Mex.astype(int)
```

```python
intervalos_Mex = range(1950, max(x_Mex)+2) #calculamos los extremos de los intervalos
año_Mex = list(range(1950, 2024, 10))
# se le quito el +2
plt.style.use('ggplot')
#plt.style.use('fivethirtyeight')
plt.hist(x=x_Mex, bins=intervalos_Mex, color='red', rwidth=0.85)
plt.title('Reportes de Entomopatógenos de México en GBIF')
plt.xlabel('Año')
plt.ylabel('Frecuencia')
plt.xticks(año_Mex,rotation=45)
plt.savefig("Histograma_Mex.png",dpi=300)
```

```python
# Para variables discretas
pd.value_counts(df_Mex['species'])
```

```python
100 * df_Mex['species'].value_counts() / len(df_Mex['species'])
```

```python
plot = (pd.value_counts(df_Mex['species'])).plot(
kind='bar', title='Especies de Entomopatógenos de México en GBIF')
plt.savefig("Especies_Mex.png",dpi=300)
```

# Guardar y crear archivos csv

```python
# si es un data frame
# df_numeros.to_csv('numeros.csv', header=False, index=False)

# si es un array
Filo = df_Mex['phylum'].array
Clase = df_Mex['class'].array
Orden = df_Mex['order'].array
Familia = df_Mex['family'].array
Genero = df_Mex['genus'].array
Especie = df_Mex['species'].array

#Para la localidad
Loc1 = df_Mex['stateProvince'].array
Loc2 = df_Mex['locality'].array

Estados = ('Aguascalientes', 'Baja California', 'Baja California Sur', 'Campeche', 'Coahuila', 'Colima', 'Chiapas', 'Chihuahua', 'Durango', 'Distrito Federal', 'Guanajuato', 'Guerrero', 'Hidalgo', 'Jalisco', 'Mexico', 'Michoacan', 'Morelos', 'Nayarit', 'Nuevo Leon', 'Oaxaca', 'Puebla', 'Queretaro', 'Quintana Roo', 'San Luis Potosi', 'Sinaloa', 'Sonora', 'Tabasco', 'Tamaulipas', 'Tlaxcala', 'Veracruz', 'Yucatán', 'Zacatecas')
```

```python
# Creacción de un array que contenga la localidad
Localidad = []
i = 0
while i<np.size(Loc1):
    if Loc1[i] in Estados:
        Localidad.append(Loc1[i])
    else:
        Localidad.append(Loc2[i])
    i=i+1
```

```python
#Construcción de un data frame con los parametros extraidos
df_bd = pd.DataFrame()
df_bd['Filo'] = Filo
df_bd['Clase'] = Clase
df_bd['Orden'] = Orden
df_bd['Familia'] = Familia
df_bd['Genero'] = Genero
df_bd['Especie'] = Especie
df_bd['Localidad'] = Localidad

df_bd
```

```python
# Guadar el data frame en un archivo csv
df_bd.to_csv('EspeciesMexLite.csv')
```
